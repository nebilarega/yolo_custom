```
import shutil
```

```
for names in os.listdir('/content/Training/OIDv4_ToolKit/OID/Dataset/train/Aircraft_Weapon'):
  file_paths = os.path.join('/content/Training/OIDv4_ToolKit/OID/Dataset/train/Aircraft_Weapon',names)
  if os.path.exists(file_paths):
    if not os.path.isdir(file_paths):
      shutil.copy(file_paths, '/content/darknet/data/obj')
```

```
image_paths = '/content/darknet/data/obj'
image_names = [names for names in os.listdir(image_paths)]
with open('/content/darknet/data/train.txt', 'w') as f:
  for file_name in image_names:
    if file_name.endswith('jpg'):
      written_name = image_paths +'/'+ file_name
      f.write(written_name)
      f.write('\n')
```
